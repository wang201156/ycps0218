```python
import pandas as pd
df=pd.read_csv("customer_contact_status.csv")
df
```

```python
df_le=df
df_ohe=df
```

將 學歷、連絡方式及訂購 使用 OrdinalEncoder 進行 Label Encofing

```python
df["學歷"].unique(), df["連絡方式"].unique(), df["訂購"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[["博士", "大學", "高中", "高職", "專科", "國中", "小學"], ['市話', '手機'], ["頻繁", "偶爾", "一次", "從未"]])
df_le[["學歷","連絡方式","訂購"]] = encoder.fit_transform(df_le[["學歷","連絡方式","訂購"]])
df_le
```

將 工作及婚姻 使用 OneHotEncoder 進行 One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')
# 使用 fit_transform 直接對多個欄位進行編碼
encoded_data = encoder.fit_transform(df_ohe[["工作","婚姻"]])
# 取得編碼後的欄位名稱
encoded_columns = encoder.get_feature_names_out(["工作","婚姻"])
# 建立新的 DataFrame 來儲存編碼後的資料
encoded_df_ohe = pd.DataFrame(encoded_data, columns=encoded_columns)
# 將編碼後的資料與原始資料合併
df_ohe = pd.concat([df_ohe, encoded_df_ohe], axis=1)
df_ohe.drop(columns=["工作","婚姻"])
df_ohe
```
