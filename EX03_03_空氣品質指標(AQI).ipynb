```python
# 讀取網路上的 CSV 檔
import pandas as pd
url = 'https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=9e565f9a-84dd-4e79-9097-d403cae1ea75&limit=1000&sort=ImportDate%20desc&format=CSV'
df = pd.read_csv(url)
df.head()
```

留下需要的欄位（'sitename', 'county','pollutant', 'status', 'aqi'）

```python
selected_columns = ['sitename', 'county', 'pollutant', 'status', 'aqi']
new_df = df[selected_columns]
new_df
```

請找出除了空氣品質不良的測站（aqi > 100）

```python
condition = (new_df['aqi'] > 100 )
new_df[condition]
```

查詢空氣污染指標物（pollutant）有那些？

```python
print("pollutant:", new_df['pollutant'].unique() )
```

請找出高雄市的測站資料

```python
condition = new_df['county'] == "高雄市"
new_df[condition]
```

計算各縣市的平均 AQI

```python
df_gp = df.groupby('county')
df_gp['aqi'].mean().sort_values(ascending=False)
```
