```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}

df = pd.DataFrame(data)
df

```

```python
avg_sal=df['Salary'].median()
df_hs=df
df_hs['High Salary'] = df_hs['Salary'].apply(lambda x: 'True' if x > df['Salary'].quantile(0.75) else 'False')
df_hs
```

```python
df_hs_filtered = df_hs[df_hs['High Salary'] == 'False']
df_hs_filtered
```

```python
new_data = [
    {'ID': 111, 'Name': 'Kevin', 'Age': 28, 'City': 'Austin', 'Salary': None, 'Department': 'IT'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'},
    {'ID': 112, 'Name': 'Lily', 'Age': 26, 'City': 'Seattle', 'Salary': 55000, 'Department': 'Marketing'}
]
df = pd.concat([df, pd.DataFrame(new_data)], ignore_index=True)
df
```

```python
df_gp = df.groupby('Department')
df_gp['Salary'].mean()
```

```python
#df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Salary'] = df_gp['Salary'].transform(lambda x: x.fillna(x.mean()))
df
```

```python
df_cleaned = df.drop_duplicates(keep='last')
df_cleaned
```

 'NewSalary' 500 'RankSalary' 

```python
df_cleaned['NewSalary'] =df_cleaned['Salary'].apply(lambda x: x+500 if x < df['Salary'].mean()  else x)
df_cleaned['RankSalary'] = df_cleaned['Salary'].rank(ascending=False, method='min')
df_cleaned['RankSalary'] = df_cleaned['RankSalary'].astype(int)
df_cleaned
```
