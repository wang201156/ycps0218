```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tips.dtypes
```

tips 中數值資料的相關分析

```python
corr=tips[["total_bill","tip","size"]].corr()
corr
```

tips 中數值資料的相關分析熱圖

```python
sns.heatmap(corr,cmap='Blues',annot=True)
```

顯示每筆資料概覽總金額和小費的關係

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Total Bill vs Tip")
plt.show()
```

總金額和小費的關係中，是否抽菸的情形如何？

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker" )
plt.title("Total Bill vs Tip")
plt.show()
```

總金額和小費的關係中，考慮是否抽菸及用餐時段的情形如何？

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker",style="time" )
plt.title("Total Bill vs Tip")
plt.show()
```

星期某天用餐數量及消費金額的關係（使用 catplot()）

```python
sns.catplot(data=tips,x="day", y="total_bill")
plt.show()
```

星期某天用餐數量及消費金額，加上性別的關係

```python
sns.catplot(data=tips,x="day", y="total_bill",hue="sex",kind="swarm")
plt.show()
```

使用箱形圖顯示不同性別，小費的高低情形（kind=box）

```python
sns.boxplot(data=tips, x="sex", y="tip",hue="sex")
plt.show()
```
